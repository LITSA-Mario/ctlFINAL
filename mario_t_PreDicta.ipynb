***Text entry is located at the bottom**
* You will be prompted to run a search on the Google sheet
* Once the results come up, enter the leftmost number which coresponds your desired entry.
* PreDicta will use Chat GPT to return a case analysis to the last column in your google sheet!

```python
import re
import openai

def search_rows_partial_entry(partial_entry):
    all_data = worksheet.get_all_values()
    matching_rows = {}
    gpt_column_index = -1

    # Find the index of the column containing the header "GPT analysis"
    existing_columns = all_data[0]
    if "GPT analysis" in existing_columns:
        gpt_column_index = existing_columns.index("GPT analysis")

    for i, row in enumerate(all_data):
        if i == 0: #skip first row
            continue

        found = False
        for j, cell in enumerate(row):
            if j != gpt_column_index:  # Exclude the GPT analysis column
                if partial_entry.lower() in cell.lower():
                    if cell.lower().startswith(partial_entry.lower()):
                        found = True
                        break

        if found:
            matching_rows[i + 1] = row

    return [(key, value) for key, value in matching_rows.items()]

# This Function will present relevant matches and request user imput to select one of the options
def choose_relevant_row(matching_rows):
    non_header_matching_rows = [row for row in matching_rows if row[0] != 1]

    if len(non_header_matching_rows) > 0:
      print("Matching clients found:")
      row_number = 1
      for row_num, row_data in non_header_matching_rows:
         # if row_num != 1:
          print(f"{row_number}. {row_data} - Row {row_num}")  # Display matching rows to choose from
          row_number += 1


      #for i, (row_num, row_data) in enumerate(non_header_matching_rows, start=1):

    print("\nEnter the leftmost number which corresponds with the entry you would like to select: ")
    selected_row_index = int(input("\n> "))

    if 1 <= selected_row_index < len(non_header_matching_rows):
        selected_row = non_header_matching_rows[selected_row_index - 1] #non_header_matching_rows[selected_row_index]
        print(f"\nYou selected:")
        print(', '.join(map(str, selected_row[1:])))
        #print(f"Row {selected_row[0]}: {', '.join(map(str, selected_row[1:]))}")
        return selected_row
    else:
        print("Invalid selection. Exiting.")
        return None

def identify_accident_type(row_data):
    description = ' '.join(row_data).lower()

    # Regular expressions defined for different accident types
    car_crash_patterns = r'\b(car|auto|automobile|vehicle)\b'
    workplace_incident_patterns = r'\b(workplace|work\s*comp|work\s*accident|job\s*accident)\b'

    # Basic check for patterns related to specific accident types in the description (to be further developed following fall 2023 semester)
    if re.search(car_crash_patterns, description):
        return "car crash"
    elif re.search(workplace_incident_patterns, description):
        return "workplace incident"
    else:
        return "other"

def generate_detailed_liability_analysis(row_data):
    # Find the index of the column containing the header "GPT analysis"
    gpt_column_index = -1  # Initialize with an invalid index
    for index, value in enumerate(row_data):
        if value.lower() == "gpt analysis":
            gpt_column_index = index
            break

    # Determine the accident type using the updated identify_accident_type function
    accident_type = identify_accident_type(row_data)


    # The following section of code is to assure that the column added to the google sheet by this program ("GPT analysis") remains "unsearched" when performing a search of the database
    # This is to exclude the content in the GPT analysis column from row_data to minimize the likelyhood of unecessary results coming up on a partial search.
    row_data_without_gpt = [data for index, data in enumerate(row_data) if index != gpt_column_index]

    # GPT prompts RE: liability & specific automobile accident type
    if accident_type == "car collision":
        prompt = "Scenario: An incident report details a car collision. \nDescription: " + ', '.join(row_data) + "\nBased on this information, assess the liability in this car collision scenario, considering the actions of drivers and road conditions."

    elif accident_type == "car-to-pedestrian incident":
        prompt = "Scenario: An incident report details a car-to-pedestrian incident. \nDescription: " + ', '.join(row_data) + "\nBased on this information, assess the liability in this car-to-pedestrian incident scenario, considering the driver's actions and pedestrian's presence or behavior."

    else:
        prompt = "Scenario: An incident report details an incident. \nDescription: " + ', '.join(row_data) + "\nBased on this information, assess the liability in this scenario, considering the specifics of the incident and the circumstances involved."

    # Call OpenAI
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=300  # max_tokens determines complexity of the response
    )

    return response.choices[0].text.strip()

# Used to get existing columns
def add_gpt_analysis_column():
    existing_columns = worksheet.row_values(1)

    if "GPT analysis" not in existing_columns:
        worksheet.add_cols(1)  # Add a new "GPT analysis" column at the end
        worksheet.update_cell(1, len(existing_columns) + 1, "GPT analysis")  #Header for the new column
        print("Added 'GPT analysis' column to the sheet.")
    else:
        print("'GPT analysis' column already exists.")

def write_liability_to_sheet(row_num, liability_output):
    add_gpt_analysis_column()
    existing_columns = worksheet.row_values(1)
    gpt_column_index = existing_columns.index("GPT analysis") + 1

    #adjusted_row_num = row_num if row_num <= 1 else row_num - 1

    worksheet.update_cell(row_num, gpt_column_index, liability_output)


def search_rows_partial_entry(partial_entry):
    all_data = worksheet.get_all_values()
    matching_rows = {}
    for i, row in enumerate(all_data):
        for cell in row:
            if partial_entry.lower() in cell.lower():
                matching_rows[i+1] = row  # Store row number and data

    # This removes duplicate rows based on their values
    unique_matching_rows = {tuple(value): key for key, value in matching_rows.items()}

    return [(key, value) for value, key in unique_matching_rows.items()]

# Entering User side!
partial_entry = input("Please submit an entry to search: ")

matching_rows = search_rows_partial_entry(partial_entry)
if matching_rows:
    selected_row = choose_relevant_row(matching_rows)
    if selected_row:
        row_num, row_data = selected_row[0], selected_row[1]

        # Generate detailed liability analysis using Chat GPT
        liability_analysis = generate_detailed_liability_analysis(row_data)
        print("\nDetailed Liability Analysis:")
        print(liability_analysis)


        write_liability_to_sheet(row_num, liability_analysis)
else:
    print("No matching rows found.")
```
